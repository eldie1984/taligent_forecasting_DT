# Average Liquor Prices Per County Table

This notebook creates the average liquor prices per county table with the following attributes:
- County
- Year
- Month
- Category
- Average price per liter

```python
import pandas as pd
import numpy as np

print("Libraries imported successfully.")
```

```python
# Load the processed data
df = pd.read_pickle('liquor_sales_processed.pkl')

print(f"Dataset shape: {df.shape}")
```

```python
# Calculate price per liter for each transaction
# Price per liter = sales_dollars / sales_liters
# Handle division by zero
df['price_per_liter'] = df['sales_dollars'] / df['sales_liters']
df['price_per_liter'] = df['price_per_liter'].replace([np.inf, -np.inf], np.nan)

# Check for any remaining issues
print(f"Number of records with valid price per liter: {df['price_per_liter'].notna().sum()}")
print(f"Number of records with invalid price per liter: {df['price_per_liter'].isna().sum()}")
```

```python
# Create average liquor prices per county table
# Group by county, year, month, and category
avg_prices = df.groupby(['county_name', 'year', 'month', 'category_name']).agg({
    'price_per_liter': 'mean'
}).reset_index()

# Rename columns to match requirements
avg_prices.columns = ['County', 'Year', 'Month', 'Category', 'Average price per liter']

# Sort by county, year, month, and category
avg_prices = avg_prices.sort_values(['County', 'Year', 'Month', 'Category'], ascending=[True, True, True, True])

print(f"Average prices table shape: {avg_prices.shape}")
print(f"\nFirst 10 rows:")
avg_prices.head(10)
```

```python
# Display summary statistics
print("\nSummary statistics:")
print(f"Number of counties: {avg_prices['County'].nunique()}")
print(f"Number of categories: {avg_prices['Category'].nunique()}")
print(f"Number of years: {avg_prices['Year'].nunique()}")
print(f"Number of months: {avg_prices['Month'].nunique()}")
print(f"Total records: {len(avg_prices)}")
```

```python
# Display overall price statistics
print("\nAverage price per liter statistics:")
print(avg_prices['Average price per liter'].describe())
```

```python
# Display categories with highest average prices
category_avg_prices = avg_prices.groupby('Category')['Average price per liter'].mean().sort_values(ascending=False)
print("\nTop 10 categories by average price per liter:")
print(category_avg_prices.head(10))
```

```python
# Display counties with highest average prices
county_avg_prices = avg_prices.groupby('County')['Average price per liter'].mean().sort_values(ascending=False)
print("\nTop 10 counties by average price per liter:")
print(county_avg_prices.head(10))
```

```python
# Check for price trends over time
yearly_avg_prices = avg_prices.groupby('Year')['Average price per liter'].mean()
print("\nAverage price per liter by year:")
print(yearly_avg_prices)
```

```python
# Save the average prices table
avg_prices.to_csv('average_liquor_prices_county.csv', index=False)
print("\nAverage prices table saved to average_liquor_prices_county.csv")
```

```python
# Also save as pickle for use in other notebooks
avg_prices.to_pickle('average_prices_county.pkl')
print("Average prices table saved to average_prices_county.pkl")
```

## Summary

Created average liquor prices per county table with:
- County: County name
- Year: Year of sales
- Month: Month of sales
- Category: Category name
- Average price per liter: Mean price per liter for each county-year-month-category combination

This table provides insights into:
- Price variations across counties
- Seasonal price trends
- Category-specific pricing patterns
- Regional price differences

**Note:** The average price per liter is calculated as sales_dollars / sales_liters for each transaction, then averaged across all transactions in each county-year-month-category group.
