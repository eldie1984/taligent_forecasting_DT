# Store-Level Liquor Sales Table

This notebook creates the store-level liquor sales table with the following attributes:
- Store
- Product
- Total Sales
- Year
- Month

```python
import pandas as pd
import numpy as np

print("Libraries imported successfully.")
```

```python
# Load the processed data
df = pd.read_pickle('liquor_sales_processed.pkl')

print(f"Dataset shape: {df.shape}")
```

```python
# Create store-level liquor sales table
# Group by store (store_no + store_name), product (im_desc), year, and month
store_sales = df.groupby(['store_no', 'store_name', 'im_desc', 'year', 'month']).agg({
    'sales_dollars': 'sum'
}).reset_index()

# Create a combined store identifier
store_sales['Store'] = store_sales['store_no'].astype(str) + ' - ' + store_sales['store_name']

# Select and rename columns to match requirements
store_sales = store_sales[['Store', 'im_desc', 'sales_dollars', 'year', 'month']]
store_sales.columns = ['Store', 'Product', 'Total Sales', 'Year', 'Month']

# Sort by store, year, month, and total sales
store_sales = store_sales.sort_values(['Store', 'Year', 'Month', 'Total Sales'], ascending=[True, True, True, False])

print(f"Store-level sales table shape: {store_sales.shape}")
print(f"\nFirst 10 rows:")
store_sales.head(10)
```

```python
# Display summary statistics
print("\nSummary statistics:")
print(f"Number of stores: {store_sales['Store'].nunique()}")
print(f"Number of products: {store_sales['Product'].nunique()}")
print(f"Number of years: {store_sales['Year'].nunique()}")
print(f"Number of months: {store_sales['Month'].nunique()}")
print(f"Total records: {len(store_sales)}")
```

```python
# Display top stores by total sales
top_stores = store_sales.groupby('Store')['Total Sales'].sum().sort_values(ascending=False)
print("\nTop 10 stores by total sales:")
print(top_stores.head(10))
```

```python
# Display sales distribution by month
monthly_sales = store_sales.groupby('Month')['Total Sales'].sum()
print("\nTotal sales by month:")
print(monthly_sales)
```

```python
# Display top products by total sales
top_products = store_sales.groupby('Product')['Total Sales'].sum().sort_values(ascending=False)
print("\nTop 10 products by total sales:")
print(top_products.head(10))
```

```python
# Save the store-level sales table
store_sales.to_csv('store_level_liquor_sales.csv', index=False)
print("\nStore-level sales table saved to store_level_liquor_sales.csv")
```

```python
# Also save as pickle for use in other notebooks
store_sales.to_pickle('store_level_sales.pkl')
print("Store-level sales table saved to store_level_sales.pkl")
```

## Summary

Created store-level liquor sales table with:
- Store: Store number and name combined
- Product: Product description (im_desc)
- Total Sales: Sum of sales_dollars for each store-product-year-month combination
- Year: Year of sales
- Month: Month of sales

This table provides granular sales data at the store level, allowing for:
- Store performance analysis
- Product popularity by store
- Seasonal trends at the store level
- Inventory planning by store
