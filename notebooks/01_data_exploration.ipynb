# Iowa Liquor Sales Data Exploration

This notebook explores the Iowa Liquor Sales dataset to understand its structure and prepare for data engineering tasks.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 50)

print("Libraries imported successfully.")
```

```python
# Load the Iowa Liquor Sales data
df = pd.read_csv('../data/annual_population_estimates_for_state_of_iowa_cities_and_counties_1101_rows.csv')

print(f"Dataset shape: {df.shape}")
print(f"\nDataset size: {df.memory_usage(deep=True).sum() / 1024 / 1024:.2f} MB")
```

```python
# Display first few rows
df.head()
```

```python
# Display column information
print("Column names:")
print(df.columns.tolist())
```

```python
','.join(df.columns.tolist())
```

```python
# Display data types
print("\nData types:")
print(df.dtypes)
```

```python
# Display basic statistics
print("\nBasic statistics:")
df.describe()
```

```python
# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Convert date column to datetime
df['ordered_on'] = pd.to_datetime(df['ordered_on'])

# Extract year and month
df['year'] = df['ordered_on'].dt.year
df['month'] = df['ordered_on'].dt.month

print(f"Date range: {df['ordered_on'].min()} to {df['ordered_on'].max()}")
print(f"Years in dataset: {sorted(df['year'].unique())}")
```

```python
# Explore unique values in key categorical columns
print(f"Number of unique counties: {df['county_name'].nunique()}")
print(f"Number of unique stores: {df['store_no'].nunique()}")
print(f"Number of unique categories: {df['category_name'].nunique()}")
print(f"Number of unique products: {df['im_desc'].nunique()}")
```

```python
# Display top counties by sales
county_sales = df.groupby('county_name')['sales_dollars'].sum().sort_values(ascending=False)
print("\nTop 10 counties by total sales:")
print(county_sales.head(10))
```

```python
# Display top categories by sales
category_sales = df.groupby('category_name')['sales_dollars'].sum().sort_values(ascending=False)
print("\nTop 10 categories by total sales:")
print(category_sales.head(10))
```

```python
# Sales by year
yearly_sales = df.groupby('year')['sales_dollars'].sum()
print("\nTotal sales by year:")
print(yearly_sales)
```

```python
# Save the processed dataframe for use in other notebooks
df.to_pickle('liquor_sales_processed.pkl')
print("\nProcessed data saved to liquor_sales_processed.pkl")
```
