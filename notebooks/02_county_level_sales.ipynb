# County-Level Liquor Sales Table

This notebook creates the county-level liquor sales table with the following attributes:
- County
- Product
- Total Sales
- Year
- County population (placeholder - requires Iowa Census data)

```python
import pandas as pd
import numpy as np

print("Libraries imported successfully.")
```

```python
# Load the processed data
df = pd.read_pickle('liquor_sales_processed.pkl')

print(f"Dataset shape: {df.shape}")
```

```python
# Create county-level liquor sales table
# Group by county, product (im_desc), and year
county_sales = df.groupby(['county_name', 'im_desc', 'year']).agg({
    'sales_dollars': 'sum'
}).reset_index()

# Rename columns to match requirements
county_sales.columns = ['County', 'Product', 'Year', 'Total Sales']

# Sort by county, year, and total sales
county_sales = county_sales.sort_values(['County', 'Year', 'Total Sales'], ascending=[True, True, False])

print(f"County-level sales table shape: {county_sales.shape}")
print(f"\nFirst 10 rows:")
county_sales.head(10)
```

```python
# Display summary statistics
print("\nSummary statistics:")
print(f"Number of counties: {county_sales['County'].nunique()}")
print(f"Number of products: {county_sales['Product'].nunique()}")
print(f"Number of years: {county_sales['Year'].nunique()}")
print(f"Total records: {len(county_sales)}")
```

```python
# Note: County population data would need to be added from Iowa Census data
# This is a placeholder for the county population column
# To complete this, you would need to:
# 1. Load Iowa Census population data
# 2. Merge with county_sales on County and Year
# 3. Add the 'County population' column

# For now, we'll add a placeholder column
county_sales['County population'] = np.nan

print("\nNote: County population column added as placeholder.")
print("To populate this column, merge with Iowa Census population data.")
```

```python
# Display top counties by total sales
top_counties = county_sales.groupby('County')['Total Sales'].sum().sort_values(ascending=False)
print("\nTop 10 counties by total sales:")
print(top_counties.head(10))
```

```python
# Display top products by total sales
top_products = county_sales.groupby('Product')['Total Sales'].sum().sort_values(ascending=False)
print("\nTop 10 products by total sales:")
print(top_products.head(10))
```

```python
# Save the county-level sales table
county_sales.to_csv('county_level_liquor_sales.csv', index=False)
print("\nCounty-level sales table saved to county_level_liquor_sales.csv")
```

```python
# Also save as pickle for use in other notebooks
county_sales.to_pickle('county_level_sales.pkl')
print("County-level sales table saved to county_level_sales.pkl")
```
